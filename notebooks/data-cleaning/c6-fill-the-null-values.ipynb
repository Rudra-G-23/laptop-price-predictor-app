```python
import polars as pl

expression = r"(\d+),(\d+)"  # Simplified regex for extracting digits around a comma

dl.with_columns(
    # Extract the first capture group (the digits before the comma)
    pl.col('mrp').str.extract(expression, 1).alias("mrp_extracted")
)
```

```python
expression = "(?P<digit>(\d{1,})(?:,)(\d{1,}))"
```

```python
dl2 = dl.with_columns(
    pl.col("mrp")
    .str.replace_all(r"[^\d]", "")
)
```

```python
dl.with_columns(
    pl.col("mrp")
      .str.replace_all(r'[₹,",]', "")   # remove ₹ , and "
      .cast(pl.Int64)
)
```

```python
dl.with_columns(
    pl.col("mrp")
    .str.replace_all(r"""[",₹"]""", "")
)
```

```python
dl2['mrp']
```

```python
dl.with_columns(
    pl.col("mrp")
    .str.replace_all(r"""[₹,"]""", "")
    .cast(pl.Int16)
)
```

```python
dl.with_columns(
    pl.col('mrp')
    .str.contains("(?P<digit>(\d{1,})(?:,)(\d{1,}))")
)
```

```python
forward_fill_cols = ['rating', 'total_rating', 'next_month_bought', 'offer', 'special_features']

dl = dl.with_columns(
    pl.col(forward_fill_cols).fill_null(strategy='forward')
)

dl.null_count()
```
